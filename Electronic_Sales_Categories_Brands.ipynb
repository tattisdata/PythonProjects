```python
import pandas as pd
```

```python
df = pd.read_csv('electronics_bestymd.csv', sep=',', header=0)
df
```

```python
df.category.value_counts()
```

```python
df.category.value_counts(normalize=True)
```

```python
df.brand.value_counts()
```

```python
df.pivot_table(index='brand', columns='category', values='user_id', aggfunc='count', fill_value=0, margins=True)
```

```python
data_res_cat = df.groupby('category')[['item_id']].count().sort_values(by='item_id', ascending=False).head(10)
data_res_cat
```

```python
data_res_br = df.groupby('brand')[['item_id']].count().sort_values(by='item_id', ascending=False).head(20)
data_res_br
```

```python
res_category = df.groupby('category')[['item_id']].count()
res_category
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [20, 14]
plt.barh(res_category.index, res_category.item_id, color = "green")
plt.yticks(range(0, 11))
plt.xticks(range(0, int(round(res_category.max()[0])), 40000))
plt.ylabel('Number of items sold')
plt.xlabel('Category')
plt.title('Best Sales Category')    
plt.show()
```

```python
res_brand = df.groupby('brand')[['item_id']].count()
res_brand
```

```python
res_brand.info()
```

```python
res_brandv1 = res_brand.drop(['Other'])
```

```python
plt.rcParams['figure.figsize'] = [20, 26]
plt.barh(res_brandv1.index, res_brandv1.item_id, color = "orange")
plt.yticks(range(0, 50))
plt.xticks(range(0, int(round(res_brandv1.max()[0])), 2000))
plt.ylabel('Number of items sold')
plt.xlabel('Brand')
plt.title('Best Sales Brand')    
plt.show()
```
